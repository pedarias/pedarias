# Scatter Plot Game

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random

def scatter_plot():
  xmin = -20
  xmax = 20
  ymin = -20
  ymax = 20

  for num in range(0,3):
    print("Round: ", num+1, "/3")
    fig, ax = plt.subplots()
    plt.axis([xmin, xmax, ymin, ymax])
    plt.plot([xmin, xmax], [0,0], 'black')
    plt.plot([0,0], [ymin, ymax], 'black')
    ax.set_xlabel("x values")
    ax.set_ylabel("y values")
    ax.set_title("Scatter Plot Game")
    ax.set_xticks(np.arange(xmin, xmax, 2))
    ax.set_yticks(np.arange(ymin, ymax, 2))
    ax.grid(True)
    randomx = random.randint(-20, 20)
    randomy = random.randint(-20, 20)
    plt.plot([randomx], [randomy], 'ro')
    plt.show()
    answer = input("Enter X and Y coordinates seperated by a comma: ")
    answer = answer.split(',')
    xans = int(answer[0])
    yans = int(answer[1])
    if xans == randomx and yans == randomy:
      print("Correct!")
    else:
      print("Incorrect, the answer was (", randomx, ",", randomy, ")")
scatter_plot()
```

# Algerba Practice Game

```python
import sympy
from sympy import symbols
from sympy.solvers import solve
import numpy as np
from numpy import random

def algerba():
  eq = None
  x = symbols('x')

  genorator = random.randint(100)
  if genorator >= 50:
    num = random.randint(100)
    if genorator <= 75:
      print("Equation: y = x -", num)
      eq = x - num
    else:
      print("Equation: y = x +", num)
      eq = x + num
  else:
    num1 = random.randint(100)
    num2 = random.randint(100)
    if genorator <= 12:
      print("Equation: y = ", num1, "* x +", num2)
      eq = num1*x + num2
    elif genorator > 12 and genorator <= 24:
      print("Equation: y = -", num1, "* x +", num2)
      eq = -num1*x + num2
    elif genorator > 24 and genorator <= 36:
      print("Equation: y = ", num1, "* x -", num2)
      eq = num1*x -num2
    else:
      print("Equation: y = -", num1, "* x -", num2)
      eq = -num1*x - num2
  answer = input("Solve for X: ")
  if "/" in answer:
    nd = answer.split("/")
    n = float(nd[0])
    d = float(nd[1])
    answer = n/d
  else:
    answer = float(answer)
  solution = solve(eq, x)
  sol = str(solution[0])
  if "/" in sol:
    nd = sol.split("/")
    n = float(nd[0])
    d = float(nd[1])
    sol = n/d
  else:
    sol = float(sol)
  if sol == answer:
    print("Correct!")
  else:
    print("Incorrect, answer is", solution[0])
algerba()
```

# Projectile Game

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import *
from numpy import random

def projectile():
  targetx = None
  targety = None
  wallx = random.randint(1,18)
  wally = random.randint(1,18)
  def f(a, b, c):
    xmin = 0
    xmax = 20
    ymin = 0
    ymax = 20
    points = 10*xmax-xmin
    x = np.linspace(xmin, xmax, points)

    plt.axis([xmin,xmax,ymin,ymax])

    plt.plot([wallx, wallx], [0, wally], '80r')

    targetx = wallx
    targety = wally + 1
    plt.plot([targetx], [targety], 'go')

    y = a*x**2 + b*x + c
    plt.plot(x, y)
    plt.show()
    return targetx, targety

  tracker = True

  print("Try to enter the correct a, b, and c values to get the vertex on the red dot!")
  while tracker is True:
    print("Enter: 444 in a to exit program if needed!")
    a = int(input("a value: "))
    b = int(input("b value: "))
    c = int(input("c value: "))
    if a == 444:
      tracker = False
    vx = -b/(2*a)
    vy = a*(vx**2) + b*vx + c
    targetx, targety = f(a,b,c)
    print("Target vertex: (", targetx, ",", targety, ")")
    print("Projection vertex: (", vx, ",", vy, ")")
    if vx == targetx and vy == targety:
      tracker = False
      print("YAAAAA! You got it, good job!")
projectile()
```
